#Magnum Opus

<h3><ul><li>Bag-of-Words (Document corpus)</ul></li><h3>

```python
import requests
import bs4
url=['https://en.wikipedia.org/wiki/Web_mining','https://en.wikipedia.org/wiki/Data_mining',
     'https://en.wikipedia.org/wiki/Artificial_intelligence','https://en.wikipedia.org/wiki/Machine_learning',
     'https://en.wikipedia.org/wiki/Mining']
fdata=[]
for x in range(5):
    res=requests.get(url[x])
    soup=bs4.BeautifulSoup(res.text,'html')
    data=soup.select('p')
    newdata=""
    for i in range(len(data)):
         newdata=newdata+data[i].getText()
    fdata.append(newdata)
print(fdata)
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize,sent_tokenize
stopword=(stopwords.words('english'))
filtered_text=[]
for x in range(5):
    filtered_text.append([])
    word_tokens=word_tokenize(fdata[x])
    for w in word_tokens:
        if w not in stopword and w.isalnum()==True:
            filtered_text[x].append(w.lower())

from nltk.stem import LancasterStemmer
stemmer=LancasterStemmer()
term_freq=[]
all_terms={}
total=[]
for x in range(5):
    for y in range(len(filtered_text[x])):
        filtered_text[x][y]=stemmer.stem(filtered_text[x][y])
        all_terms[filtered_text[x][y]]=0
for x in range(5):
    total.append(0)
    term_freq.append({})
    for term in all_terms.keys():
        term_freq[x][term]=0
    for term in filtered_text[x]:
        term_freq[x][term]+=1
        total[x]+=1

import pandas as pd
data_disp_bow={'Doc-ID':['D1','D2','D3','D4','D5']}
col=['Doc-ID']

for x in all_terms.keys():
    col.append(x)
    data_disp_bow[x]=[]
    for y in range(5):
        data_disp_bow[x].append(term_freq[y][x])
df = pd.DataFrame(data_disp_bow,columns=col)
df
```

<h3><ul><li>TF (Document corpus)</ul></li></h3>

```python
vector=[]
for x in range(5):
    vector.append([])
    for term in all_terms.keys():
        vector[x].append(round(term_freq[x][term]/total[x],4))
print(vector)
```

<h3><ul><li>IDF (Document corpus)</ul></li></h3>

```python
import math
idf=[]
for term in all_terms.keys():
    d=0
    for x in range(5):
        if term_freq[x][term]!=0:
            d+=1
    idf.append(math.log(6/d))
print(idf)
```

<h3><ul><li>TF-IDF (Document corpus)</ul></li></h3>

```python
tfidf=[]
for x in range(5):
    tfidf.append([])
    for y in range(len(idf)):
        tfidf[x].append(vector[x][y]*idf[y])
print(tfidf)
```

<h3><ul><li>Document Similarity (Among Documents)</ul></li></h3>

```python
cosine=[]
euclidean=[]
i=0
for x in range(5):
    for y in range(x+1,5):
        cosine.append(0)
        euclidean.append(0)
        for z in range(len(tfidf[0])):
            cosine[i]+=tfidf[x][z]*tfidf[y][z]
            euclidean[i]+=pow(tfidf[x][z]-tfidf[y][z],2)
        euclidean[i]=math.sqrt(euclidean[i])
        euclidean[i]=[euclidean[i],'Doc'+str(x+1)+' & Doc'+str(y+1)]
        cosine[i]=[cosine[i],'Doc'+str(x+1)+' & Doc'+str(y+1)]
        i+=1

print('Similarity of documents based on cosine similarity')
cosine.sort()
for x in range(i-1,-1,-1):
    print(i-x,cosine[x][1])
print()
print('Similarity of documents based on euclidean distance')
euclidean.sort()
for x in range(i):
    print(x+1,euclidean[x][1])
```
